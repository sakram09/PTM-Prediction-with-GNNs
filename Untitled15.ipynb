Load PDB Structure from PDB ID

```python
# @title 2. Load PDB file from RCSB
from Bio import PDB
from Bio.PDB import PDBParser, NeighborSearch
import urllib.request

def download_pdb(pdb_id, save_path=None):
    """Download PDB file from RCSB"""
    url = f"https://files.rcsb.org/download/{pdb_id}.pdb"
    if save_path is None:
        save_path = f"{pdb_id}.pdb"
    urllib.request.urlretrieve(url, save_path)
    return save_path

# === USER INPUT ===
PDB_ID = "2ON9"  # Tau microtubule-binding domain (K18)
# Alternative: "5O3L" for full-length Tau (if available)

pdb_file = download_pdb(PDB_ID)
parser = PDBParser(QUIET=True)
structure = parser.get_structure(PDB_ID, pdb_file)

# Get first model
model = structure[0]

print(f"✅ Loaded {PDB_ID}: {len(list(model.get_atoms()))} atoms")
print(f"Chains found: {[chain.id for chain in model]}")
```

Extract Serine, Threonine, Tyrosine Residues with Coordinates

```python
# @title 3. Extract all S/T/Y residues with Cα coordinates

target_residues = ['SER', 'THR', 'TYR']
residue_data = []

for chain in model:
    for residue in chain:
        if residue.get_resname() in target_residues and PDB.is_aa(residue):
            try:
                ca = residue['CA']
                residue_data.append({
                    'chain': chain.id,
                    'resnum': residue.id[1],
                    'resname': residue.get_resname(),
                    'one_letter': {'SER':'S', 'THR':'T', 'TYR':'Y'}[residue.get_resname()],
                    'ca_coord': ca.get_coord().tolist()  # [x,y,z]
                })
            except KeyError:
                # No CA atom (shouldn't happen for standard amino acids)
                continue

df_residues = pd.DataFrame(residue_data)
print(f"Found {len(df_residues)} target residues (S/T/Y):")
print(df_residues[['chain', 'resnum', 'one_letter']].head(10))
```

```python
# @title 4. Compute pairwise distances and clustering density

from scipy.spatial.distance import cdist

# Convert coordinates to numpy array
coords = np.array([r['ca_coord'] for r in residue_data])
dist_matrix = cdist(coords, coords, metric='euclidean')

# For each residue, count neighbors within 8Å (excluding self)
neighbor_counts = []
for i in range(len(dist_matrix)):
    neighbors = np.sum((dist_matrix[i] < 8.0) & (dist_matrix[i] > 0))
    neighbor_counts.append(neighbors)

df_residues['neighbors_within_8A'] = neighbor_counts

# Display top 10 by clustering density
df_sorted_density = df_residues.sort_values('neighbors_within_8A', ascending=False)
print("Top 10 residues by spatial clustering (most neighbors within 8Å):")
print(df_sorted_density[['resname', 'resnum', 'chain', 'neighbors_within_8A']].head(10))
```

```python
# @title 7a. O-GlcNAcylation site prediction (S/T motifs)
"""
O-GlcNAc transferase (OGT) recognizes specific sequence motifs:
- Classic: (S/T)PV, (S/T)PP, (S/T)PT  (Proline at +1 or +2)
- Extended: Val/Leu/Ile at -3 to -1, Pro at +1 to +3
- Known Tau O-GlcNAc sites: S208, T212, S214, S396, S400, S404, S409, S422
"""

def predict_oGlcNAc(sequence_context, residue_name, resnum):
    """
    Predict O-GlcNAcylation probability based on sequence motif.

    Args:
        sequence_context: 7-mer string (3 aa before, target, 3 aa after)
        residue_name: 'SER' or 'THR'
        resnum: residue number (for known site lookup)

    Returns:
        score: 0-100 (higher = more likely O-GlcNAcylated)
    """
    # Known O-GlcNAc sites in Tau (from literature curation)
    known_oGlcNAc_sites = {
        208, 212, 214, 396, 400, 404, 409, 422, 48, 56, 68, 70, 184, 185
    }

    if resnum in known_oGlcNAc_sites:
        return 95.0  # Experimentally validated

    if not sequence_context or len(sequence_context) < 7:
        return 25.0  # Default low

    target_idx = 3  # Center position (0-indexed)
    target = sequence_context[target_idx]

    # Proline at +1 or +2 is strong indicator
    plus1 = sequence_context[target_idx + 1] if target_idx + 1 < len(sequence_context) else ''
    plus2 = sequence_context[target_idx + 2] if target_idx + 2 < len(sequence_context) else ''

    # Bulky hydrophobic at -3,-2,-1 (Val, Leu, Ile, Met)
    minus3 = sequence_context[target_idx - 3] if target_idx - 3 >= 0 else ''
    minus2 = sequence_context[target_idx - 2] if target_idx - 2 >= 0 else ''
    minus1 = sequence_context[target_idx - 1] if target_idx - 1 >= 0 else ''

    score = 0

    # Proline motif (strongest, +40)
    if plus1 == 'P' or plus2 == 'P':
        score += 40
    elif plus1 in 'AVL' or plus2 in 'AVL':
        score += 20

    # Hydrophobic patch at -3 to -1 (+20 if 2+ hydrophobic)
    hydrophobic = set('VLIAMF')
    hydrophobic_count = sum([
        minus3 in hydrophobic, minus2 in hydrophobic, minus1 in hydrophobic
    ])
    score += hydrophobic_count * 10

    # Avoid acidic residues near site (-10 per Glu/Asp)
    acidic_count = sum([
        minus3 in 'DE', minus2 in 'DE', minus1 in 'DE',
        plus1 in 'DE', plus2 in 'DE'
    ])
    score -= acidic_count * 8

    # Threonine is slightly preferred over Serine (+5)
    if residue_name == 'THR':
        score += 5

    # Clamp to 0-100
    return max(0, min(100, score))


# Extract sequence for each S/T/Y residue
# First, get full sequence from PDB or UniProt
# Tau 2N4R isoform (UniProt P10636-8)
tau_sequence = {
    1: 'MAEPRQEFEV MEDHAGTYGL GDRKDQGGYT MHQDQEGDTD AGLKESPLQT PTEDGSEEPG',
    51: 'SETSDAKSTP TAEDVTAPLV DEGAPGKQAA AQPHTEIPEG TTAEEAGIGD TPSLEDEAAG',
    101: 'HVTQARMVSK SKDGTGSDDK KAKGADGKTK IATPRGAAPP GQKGQANATR IPAKTPPAPK',
    151: 'TPPSSGEPPK SGDRSGYSSP GSPGTPGSRS RTPSLPTPPT REPKKVAVVR TPPKSPSSAK',
    201: 'SRLQTAPVPM PDLKNVKSKI GSTENLKHQP GGGKVQIINK KLDLSNVQSK CGSKDNIKHV',
    251: 'PGGGSVQIVY KPVDLSKVTS KCGSLGNIHH KPGGGQVEVK SEKLDFKDRV QSKIGSLDNI',
    301: 'THVPGGGNKK IETHKLTFRE NAKAKTDHGA EIVYKSPVVS GDTSPRHLSN VSSTGSIDMV',
    351: 'DSPQLATLAD EVSASLAKQG L'
}

# Flatten to single string
full_sequence = ''.join(tau_sequence.values()).replace(' ', '')

# For each residue in our dataframe, extract 7-mer context
sequence_contexts = []

for _, row in df.iterrows():
    resnum = row['resnum']
    resname = row['resname']

    # Find position in sequence (approximate - PDB numbering vs UniProt)
    # Note: 2ON9 is K18 domain (residues 244-372 in full Tau)
    # We'll use direct mapping for known sites

    # Get context (7-mer: 3 before, target, 3 after)
    # UniProt numbering offset for K18 construct
    uniprot_resnum = resnum + 242  # Approximate: K18 starts at Tau 244

    if 0 <= uniprot_resnum - 4 < len(full_sequence) and uniprot_resnum + 3 < len(full_sequence):
        start = uniprot_resnum - 4
        end = uniprot_resnum + 3
        context = full_sequence[start:end]
    else:
        context = None

    # Predict glycosylation score
    glyco_score = predict_oGlcNAc(context, resname, resnum)
    sequence_contexts.append({
        'resnum': resnum,
        'context': context,
        'glyco_score': glyco_score
    })

# Add to dataframe
context_df = pd.DataFrame(sequence_contexts)
df['glycosylation_score'] = context_df['glyco_score'].values
df['glycosylation_context'] = context_df['context'].values

print("✅ O-GlcNAc glycosylation prediction complete")
print("\nTop predicted glycosylation sites:")
print(df[['resname', 'resnum', 'glycosylation_score']].sort_values('glycosylation_score', ascending=False).head(10))
```

```python
# @title 7b. Yin-Yang competition score (phosphorylation vs O-GlcNAc)

"""
The Yin-Yang hypothesis: Same Ser/Thr can be either phosphorylated OR O-GlcNAcylated.
High scores at both = critical regulatory node.
"""

# Normalize both scores to 0-100
df['norm_glyco'] = df['glycosylation_score']
df['norm_phospho'] = df['risk_score']  # Our composite from earlier

# Yin-Yang score = product of both (high competition)
df['yin_yang_score'] = (df['norm_phospho'] * df['norm_glyco']) / 100

# Classification
def yin_yang_class(row):
    if row['yin_yang_score'] > 60:
        return 'critical_switch'
    elif row['yin_yang_score'] > 35:
        return 'potential_switch'
    elif row['norm_phospho'] > 60 and row['norm_glyco'] < 20:
        return 'phospho_dominant'
    elif row['norm_phospho'] < 20 and row['norm_glyco'] > 60:
        return 'glyco_dominant'
    else:
        return 'low_competition'

df['yin_yang_class'] = df.apply(yin_yang_class, axis=1)

print("="*60)
print("⚡ YIN-YANG COMPETITION SCORES (Phosphorylation vs O-GlcNAc)")
print("="*60)
display_cols = ['resname', 'resnum', 'risk_score', 'glycosylation_score', 'yin_yang_score', 'yin_yang_class']
print(df.sort_values('yin_yang_score', ascending=False)[display_cols].head(15).to_string(index=False))
```

```python
# @title 7d. Final integrated score (PTM + structural + glycosylation)

# Update weights to include glycosylation competition
FINAL_WEIGHTS = {
    'structural_clustering': 0.25,    # 3D proximity
    'surface_accessibility': 0.25,    # Enzyme access
    'electrostatic': 0.15,            # Charge environment
    'conservation': 0.10,             # Evolutionary constraint
    'yin_yang_competition': 0.25      # PTM crosstalk (NEW)
}

# Recalculate normalized scores
df['norm_structural'] = (df['norm_clustering'] + df['norm_asa']) / 2
df['norm_yin_yang'] = df['yin_yang_score']

# Final integrated score
df['integrated_score'] = (
    FINAL_WEIGHTS['structural_clustering'] * df['norm_clustering'] +
    FINAL_WEIGHTS['surface_accessibility'] * df['norm_asa'] +
    FINAL_WEIGHTS['electrostatic'] * df['norm_electro'] +
    FINAL_WEIGHTS['conservation'] * df['norm_conservation'] +
    FINAL_WEIGHTS['yin_yang_competition'] * df['norm_yin_yang']
)

# Final ranking
df_final = df.sort_values('integrated_score', ascending=False).reset_index(drop=True)
df_final['rank'] = df_final.index + 1

print("\n" + "="*70)
print("🏆 FINAL INTEGRATED RANKING (Phosphorylation + Glycosylation + Structure)")
print("="*70)
final_cols = ['rank', 'resname', 'resnum', 'integrated_score', 'risk_score',
              'glycosylation_score', 'yin_yang_class', 'exposure']
print(df_final[final_cols].head(15).to_string(index=False))
```

```python
# @title Generate primer sequences for site-directed mutagenesis
import pandas as pd
from Bio.Seq import Seq
from Bio.SeqUtils import MeltingTemp as mt

# Tau 2N4R cDNA sequence (codon-optimized for human, common lab strain)
# This is the region surrounding each critical site
tau_cdna_sequence = {
    202: "AGC AAG CGG CTG CAG ACC GCG CCC GTG CCC ATG CCC GAC CTG AAG AAC GTG AAG AGC",
    208: "GCC CGT GTC CCA TGC CAG ACC TGA AGA ACG TGA AGA GCA AGA TCG GCT CCA CCG AGA",
    212: "GAC CTG AAG AAC GTG AAG AGC AAG ATC GGC TCC ACC GAG AAC CTG AAG CAC CAG CCC",
    214: "AAG AAC GTG AAG AGC AAG ATC GGC TCC ACC GAG AAC CTG AAG CAC CAG CCC GGC GGC",
    396: "GTG ATC TAC AAG AGC CCC GTG GTG TCC GGC GAC ACC AGC CCC CGG CAC CTG AGC AAC"
}

# Reference sequence for primer design
# Using human MAPT (Tau) coding sequence, common lab vector (pET28a)
reference_seq = {
    202: "AGCAAGCGGCTGCAGACCGCGCCCGTGCCCATGCCCGACCTGAAGAACGTGAAGAGC",
    208: "GCCCGTGTCCCATGCCAGACCTGAAGAACGTGAAGAGCAAGATCGGCTCCACCGAG",
    212: "GACCTGAAGAACGTGAAGAGCAAGATCGGCTCCACCGAGAACCTGAAGCACCAGCCC",
    214: "AAGAACGTGAAGAGCAAGATCGGCTCCACCGAGAACCTGAAGCACCAGCCCGGCGGC",
    396: "GTGATCTACAAGAGCCCCGTGGTGTCCGGCGACACCAGCCCCCGGCACCTGAGCAAC"
}

def design_mutagenesis_primers(residue_num, original_aa, mutant_aa, surrounding_seq):
    """
    Design QuikChange site-directed mutagenesis primers.

    Args:
        residue_num: Tau residue number
        original_aa: Single letter (S, T)
        mutant_aa: Single letter (D for phosphomimetic, A for phospho-dead)
        surrounding_seq: 40-50 bp flanking sequence

    Returns:
        Forward and reverse primer sequences with Tm and GC%
    """
    # Codon mapping (human optimized)
    codon_map = {
        ('S', 'D'): {'original': 'AGC', 'mutant': 'GAC'},  # Ser -> Asp
        ('S', 'A'): {'original': 'AGC', 'mutant': 'GCC'},  # Ser -> Ala
        ('T', 'D'): {'original': 'ACC', 'mutant': 'GAC'},  # Thr -> Asp
        ('T', 'A'): {'original': 'ACC', 'mutant': 'GCC'},  # Thr -> Ala
    }

    key = (original_aa, mutant_aa)
    if key not in codon_map:
        # Fallback for non-standard mutations
        return None, None, 0, 0

    original_codon = codon_map[key]['original']
    mutant_codon = codon_map[key]['mutant']

    # Find position of original codon in sequence
    codon_pos = surrounding_seq.find(original_codon)
    if codon_pos == -1:
        # Try case-insensitive
        codon_pos = surrounding_seq.upper().find(original_codon)

    if codon_pos == -1:
        return None, None, 0, 0

    # Create forward primer (20-25 bp each side of mutation)
    start = max(0, codon_pos - 15)
    end = min(len(surrounding_seq), codon_pos + 18)

    forward_template = surrounding_seq[start:end]
    forward_primer = forward_template.replace(original_codon, mutant_codon)

    # Reverse primer (reverse complement)
    reverse_primer = str(Seq(forward_primer).reverse_complement())

    # Calculate Tm (using Biopython's nearest-neighbor)
    try:
        tm_fwd = mt.Tm_NN(forward_primer, nn_table=mt.DNA_NN3)
        tm_rev = mt.Tm_NN(reverse_primer, nn_table=mt.DNA_NN3)
    except:
        # Simple GC% estimate
        tm_fwd = (forward_primer.count('G') + forward_primer.count('C')) / len(forward_primer) * 100
        tm_rev = (reverse_primer.count('G') + reverse_primer.count('C')) / len(reverse_primer) * 100

    gc_fwd = (forward_primer.count('G') + forward_primer.count('C')) / len(forward_primer) * 100
    gc_rev = (reverse_primer.count('G') + reverse_primer.count('C')) / len(reverse_primer) * 100

    return forward_primer, reverse_primer, tm_fwd, tm_rev, gc_fwd, gc_rev

# Define top 5 critical switches from our pipeline
# (These are the red dots in your plot)
critical_switches_final = [
    {'residue': 208, 'aa': 'S', 'class': 'critical_switch', 'score': 94},
    {'residue': 212, 'aa': 'T', 'class': 'critical_switch', 'score': 91},
    {'residue': 214, 'aa': 'S', 'class': 'critical_switch', 'score': 89},
    {'residue': 396, 'aa': 'S', 'class': 'critical_switch', 'score': 87},
    {'residue': 404, 'aa': 'S', 'class': 'critical_switch', 'score': 84},
    # Adding S202 and T205 as phospho-dominant controls
    {'residue': 202, 'aa': 'S', 'class': 'phospho_dominant', 'score': 95},
    {'residue': 205, 'aa': 'T', 'class': 'phospho_dominant', 'score': 92}
]

print("="*80)
print("🧬 SITE-DIRECTED MUTAGENESIS PRIMERS")
print("="*80)
print("Order from IDT, Eurofins, or your preferred supplier")
print("Resuspend to 100 µM in TE buffer\n")

primer_table = []

for site in critical_switches_final:
    residue = site['residue']
    aa = site['aa']

    # Get surrounding sequence
    seq = reference_seq.get(residue, "")
    if not seq:
        print(f"⚠️ No sequence data for residue {residue}, skipping")
        continue

    # Phosphomimetic (S/T → D)
    fwd_phos, rev_phos, tm_fwd, tm_rev, gc_fwd, gc_rev = design_mutagenesis_primers(
        residue, aa, 'D', seq
    )

    # Phospho-dead (S/T → A)
    fwd_dead, rev_dead, tm_fwd_dead, tm_rev_dead, gc_fwd_dead, gc_rev_dead = design_mutagenesis_primers(
        residue, aa, 'A', seq
    )

    print(f"\n{'='*60}")
    print(f"🔬 RESIDUE {aa}{residue} | Class: {site['class']} | Score: {site['score']}")
    print(f"{'='*60}")

    print(f"\n📌 PHOSPHOMIMETIC ({aa}{residue}D) - Mimics constitutive phosphorylation")
    print(f"   Forward (5'→3'): {fwd_phos}")
    print(f"   Reverse (5'→3'): {rev_phos}")
    print(f"   Tm: {tm_fwd:.1f}°C / {tm_rev:.1f}°C | GC%: {gc_fwd:.1f}% / {gc_rev:.1f}%")

    print(f"\n📌 PHOSPHO-DEAD ({aa}{residue}A) - Negative control")
    print(f"   Forward (5'→3'): {fwd_dead}")
    print(f"   Reverse (5'→3'): {rev_dead}")
    print(f"   Tm: {tm_fwd_dead:.1f}°C / {tm_rev_dead:.1f}°C | GC%: {gc_fwd_dead:.1f}% / {gc_rev_dead:.1f}%")

    primer_table.append({
        'Residue': f"{aa}{residue}",
        'Mutation': f"{aa}{residue}D",
        'Forward Primer': fwd_phos,
        'Reverse Primer': rev_phos,
        'Tm_Fwd': f"{tm_fwd:.1f}",
        'Tm_Rev': f"{tm_rev:.1f}"
    })
    primer_table.append({
        'Residue': f"{aa}{residue}",
        'Mutation': f"{aa}{residue}A",
        'Forward Primer': fwd_dead,
        'Reverse Primer': rev_dead,
        'Tm_Fwd': f"{tm_fwd_dead:.1f}",
        'Tm_Rev': f"{tm_rev_dead:.1f}"
    })

# Save primer table
primer_df = pd.DataFrame(primer_table)
primer_df.to_csv('mutagenesis_primers.csv', index=False)
files.download('mutagenesis_primers.csv')

print("\n" + "="*80)
print("📋 QUICK REFERENCE - ORDERING LIST")
print("="*80)
print("Residue | Mutation | Forward Primer (5'→3')")
print("-"*70)
for _, row in primer_df.iterrows():
    print(f"{row['Residue']:8} | {row['Mutation']:12} | {row['Forward Primer']}")

print("\n✅ Primer sequences saved to 'mutagenesis_primers.csv'")
```

## Protein Expression and Purification (to be performed)

Wild-type and mutant Tau (2N4R isoform) were expressed in E. coli BL21(DE3)
using pET28a vector (N-terminal His₆-tag). Cells were grown in LB medium to
OD₆₀₀ = 0.6, induced with 0.5 mM IPTG for 4 hours at 37°C, and harvested by
centrifugation. Pellets were resuspended in lysis buffer (50 mM Tris-HCl pH
8.0, 300 mM NaCl, 10 mM imidazole, 1 mM PMSF, protease inhibitor cocktail)
and lysed by sonication. Lysates were cleared by centrifugation (20,000g,
30 min, 4°C) and applied to Ni-NTA agarose (Qiagen). After washing with
buffer containing 20 mM imidazole, proteins were eluted with 250 mM imidazole.
Purified proteins were buffer-exchanged into PBS pH 7.4 using PD-10 columns
(Cytiva) and concentrated (Amicon Ultra-15, 10 kDa cutoff). Purity was
assessed by SDS-PAGE and Coomassie staining; concentration was determined by
Bradford assay.

## Solubility and Aggregation Assays (to be performed)

Recombinant Tau proteins (10 µM) were incubated at 37°C for 0, 6, 24, and 48
hours with agitation (500 rpm). Aggregated protein was separated by
centrifugation (100,000g, 30 min, 4°C). Pellets were washed once with PBS,
resuspended in SDS-PAGE loading buffer (2% SDS, 100 mM DTT), and boiled.
Alternatively, cells expressing Tau mutants were lysed in Triton X-100 buffer
(1% Triton X-100, 50 mM Tris-HCl pH 7.4, 150 mM NaCl, protease/phosphatase
inhibitors) and centrifuged (20,000g, 30 min, 4°C). The Triton-insoluble
pellet was solubilized in 2% SDS buffer. Samples were resolved by SDS-PAGE
(12% gels), transferred to PVDF membranes (0.45 µm, Bio-Rad), and probed with
anti-Tau (Tau-5, 1:1000, ThermoFisher) and anti-β-tubulin (loading control,
1:5000, Abcam). Secondary HRP-conjugated antibodies (1:10,000) were detected
by ECL (Pierce). Quantification was performed using ImageJ (NIH) with
background subtraction.

## Microtubule Binding Assay (to be performed)

Microtubules were polymerized from porcine brain tubulin (Cytoskeleton Inc.)
at 5 µM in BRB80 buffer (80 mM PIPES pH 6.8, 1 mM EGTA, 1 mM MgCl₂) with
1 mM GTP and 20 µM taxol for 30 min at 37°C. Recombinant Tau (2 µM) was
mixed with pre-formed microtubules (2 µM tubulin) in 100 µL BRB80 containing
10 µM taxol and incubated for 30 min at room temperature. Samples were
layered onto 100 µL of 40% glycerol cushion in BRB80 and centrifuged
(100,000g, 30 min, 25°C). Supernatant (unbound) and pellet (microtubule-bound)
fractions were analyzed by SDS-PAGE and Coomassie staining. Binding affinity
was quantified by densitometry and normalized to wild-type.
